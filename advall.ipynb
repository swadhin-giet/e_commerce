```python
entered_people = set()
left_people = set()

while True:
    print("\n1. Person Enter")
    print("2. Person Leave")
    print("3. Show Status")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        name = input("Enter person name: ")

        if name in entered_people and name not in left_people:
            print(name, "is already registered and inside the event")
        else:
            entered_people.add(name)
            if name in left_people:
                left_people.remove(name)
            print(name, "entered the event")

    elif choice == "2":
        name = input("Enter person name leaving: ")

        if name in entered_people and name not in left_people:
            left_people.add(name)
            print(name, "left the event")
        else:
            print("Person not found or already left")

    elif choice == "3":
        inside_people = entered_people - left_people

        print("\nTotal registered:", len(entered_people))
        print("Total left:", len(left_people))
        print("Currently inside:", len(inside_people))
        print("People inside:", inside_people)

    elif choice == "4":
        print("Event tracking ended.")
        break

    else:
        print("Invalid choice")
```

```python
register_student=set()
rejected_student=set()

def registered_student(name):
    if name in rejected_student:
        print(f"{name} is already rejected")
    else:
        register_student.add(name)
        print(f"{name} is registered succesfully")

def reject_student(name):
    if name in register_student:
        register_student.remove(name)
    rejected_student.add(name)
    print(f"{name} is rejected succesfully")

def show():
    print("register students are :",register_student)
    print("rejected students are :",rejected_student)
    print("Total register students are :",len(register_student))
    print("Total rejected students are :",len(rejected_student))





while(True):
    print("options:")
    print("\n1. register students :")
    print("2. Reject Student")
    print("3. Show Status")
    print("4. Exit")

    c= input("enter the choice: ")
    if c=="1":
        name=input("enter the name of the student:")
        registered_student(name)
    elif c=="2":
        name=input("enter the name of the student")
        reject_student(name)
    elif c=="3":
        show()
    elif c=="4":
        break
    else:
        print("enter from the given options")
```
