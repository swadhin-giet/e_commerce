```python
import ipywidgets as widgets
from IPython.display import display, HTML
from PIL import Image
import io
import base64

# Function to load image and convert to bytes for ipywidgets.Image
def load_image_as_bytes(image_path, size=None):
    try:
        img = Image.open(image_path)
        if size:
            img = img.resize(size, Image.LANCZOS) # Use LANCZOS for high quality downsampling
        img_byte_arr = io.BytesIO()
        img.save(img_byte_arr, format='PNG') # Save as PNG for ipywidgets.Image
        return img_byte_arr.getvalue()
    except FileNotFoundError:
        print(f"Warning: Image not found at {image_path}")
        return None

# Paths to images (assuming they are in /content/)
logo_path = "/content/logo.jpeg"

# --- Widgets setup ---

# Load logo image
logo_bytes = load_image_as_bytes(logo_path, size=(90, 70))
if logo_bytes:
    logo_widget = widgets.Image(value=logo_bytes, format='png', width=90, height=70,
                                layout=widgets.Layout(margin='0 0 15px 0'))
else:
    logo_widget = widgets.HTML("<b>[Logo Missing]</b>", layout=widgets.Layout(margin='0 0 15px 0'))

# Title and subtitle
title_widget = widgets.HTML(
    value='<div style="font-family: Arial; font-size: 22px; font-weight: bold; color: #1E3D59;">GIET ERP</div>'
)
subtitle_widget = widgets.HTML(
    value='<div style="font-family: Arial; font-size: 12px; color: grey; margin-top: 5px; margin-bottom: 10px;">Student Login</div>'
)

# Email entry
email_label = widgets.HTML(value='<div style="font-family: Arial; font-size: 12px; margin-top: 10px; margin-bottom: 3px;">Email</div>')
email_entry = widgets.Text(
    placeholder='Enter your email',
    layout=widgets.Layout(width='280px', padding='4px', border='solid 1px #ccc')
)

# Password entry
password_label = widgets.HTML(value='<div style="font-family: Arial; font-size: 12px; margin-top: 12px; margin-bottom: 3px;">Password</div>')
password_entry = widgets.Text(
    placeholder='Enter your password',
    password=True, # This makes the input masked
    layout=widgets.Layout(width='280px', padding='4px', border='solid 1px #ccc')
)

# Output widget for messages
output_message = widgets.Output()

# Login function
def login_function(b):
    with output_message:
        output_message.clear_output()
        email = email_entry.value
        password = password_entry.value

        correct_email = "aps.rgp@gmail.com"
        correct_password = "110"

        if email == correct_email and password == correct_password:
            display(HTML('<p style="color: green; font-family: Arial;"><b>Login Success:</b> AAPKA SWAGAT HAIN JANAB</p>'))
        else:
            display(HTML('<p style="color: red; font-family: Arial;"><b>Login Failed:</b> Invalid Email or Password</p>'))

# Login button
login_btn = widgets.Button(
    description='LOGIN',
    button_style='info', # 'info' for a blue button, 'primary' for a bolder blue
    layout=widgets.Layout(width='200px', margin='20px 0 0 0', height='36px'),
    style=widgets.ButtonStyle(button_color='#1E3D59') # Custom background color
)
login_btn.on_click(login_function)

# Footer
footer_widget = widgets.HTML(
    value='<div style="font-family: Arial; font-size: 9px; color: grey; margin-top: 20px; padding-bottom: 10px;">© GIET University</div>'
)

# Assemble the widgets into a vertical box layout (mimicking the Tkinter frame)
login_card_content = [
    logo_widget,
    title_widget,
    subtitle_widget,
    email_label,
    email_entry,
    password_label,
    password_entry,
    login_btn,
    output_message,
    footer_widget
]

# Create the main container with styling
login_card = widgets.VBox(
    login_card_content,
    layout=widgets.Layout(
        border='1px solid #e0e0e0', # Light border
        padding='20px',
        background_color='white',
        align_items='center', # Center items horizontally
        width='360px',
        max_width='90%', # Responsive width
        margin='20px auto' # Center the card on the page with some top/bottom margin
    )
)

# Display the login card
display(login_card)
```
