```python
class PasswordManager:
    def __init__(self):
        self.passwords = {}
        self.key = 2
    def encrypt(self, password):
        e = ""
        for i in password:
            e += chr(ord(i) + self.key)
        return e
    def decrypt(self, password):
        d = ""
        for i in password:
            d += chr(ord(i) - self.key)
        return d
    def add_password(self, site, password):
        self.passwords[site] = self.encrypt(password)
        print("Password stored")
    def get_password(self, site):
        if site in self.passwords:
            print("Password:", self.decrypt(self.passwords[site]))
        else:
            print("Not found")
    def edit_password(self, site, new_password):
        if site in self.passwords:
            self.passwords[site] = self.encrypt(new_password)
            print("Password updated")
        else:
            print("Not found")
pm = PasswordManager()
while True:
    print("\n1 Add Password")
    print("2 Get Password")
    print("3 Edit Password")
    print("4 Exit")
    ch = input("Enter choice: ")
    if ch == "1":
        s = input("Enter site: ").lower()
        p = input("Enter password: ")
        pm.add_password(s, p)
    elif ch == "2":
        s = input("Enter site: ").lower()
        pm.get_password(s)
    elif ch == "3":
        s = input("Enter site: ").lower()
        np = input("Enter new password: ")
        pm.edit_password(s, np)
    elif ch == "4":
        break
    else:
        print("Invalid choice")
```

```python
class Freelancer:
    def __init__(self, name):
        self.name = name
        self.balance = 0
        self.projects = []
class Client:
    def __init__(self, name):
        self.name = name
        self.projects = []
class Project:
    def __init__(self, title, payment, freelancer, client):
        self.title = title
        self.payment = payment
        self.freelancer = freelancer
        self.client = client
        self.completed = False
class Marketplace:
    def __init__(self):
        self.freelancers = {}
        self.clients = {}
        self.projects = []
    def register_freelancer(self, name):
        self.freelancers[name] = Freelancer(name)
        print("Freelancer registered")
    def register_client(self, name):
        self.clients[name] = Client(name)
        print("Client registered")
    def assign_project(self, title, payment, freelancer_name, client_name):
        if freelancer_name in self.freelancers and client_name in self.clients:
            f = self.freelancers[freelancer_name]
            c = self.clients[client_name]
            p = Project(title, payment, f, c)
            self.projects.append(p)
            f.projects.append(p)
            c.projects.append(p)
            print("Project assigned")
        else:
            print("Freelancer or Client not found")
    def complete_project(self, title):
        for p in self.projects:
            if p.title == title and not p.completed:
                p.completed = True
                p.freelancer.balance += p.payment
                print("Payment processed to", p.freelancer.name)
                return
        print("Project not found")
    def show_freelancers(self):
        for f in self.freelancers.values():
            print(f.name, "Balance:", f.balance)
m = Marketplace()
while True:
    print("\n1 Register Freelancer")
    print("2 Register Client")
    print("3 Assign Project")
    print("4 Complete Project")
    print("5 Show Freelancers")
    print("6 Exit")
    ch = input("Enter choice: ")
    if ch == "1":
        name = input("Freelancer name: ")
        m.register_freelancer(name)
    elif ch == "2":
        name = input("Client name: ")
        m.register_client(name)
    elif ch == "3":
        title = input("Project title: ")
        payment = int(input("Payment: "))
        f = input("Freelancer name: ")
        c = input("Client name: ")
        m.assign_project(title, payment, f, c)
    elif ch == "4":
        title = input("Project title: ")
        m.complete_project(title)
    elif ch == "5":
        m.show_freelancers()
    elif ch == "6":
        break
    else:
        print("Invalid choice")
```

```python
class Customer:
    def __init__(self, name, email, phone):
        self.name = name
        self.email = email
        self.phone = phone
        self.logs = []
        self.sales = []
class CRM:
    def __init__(self):
        self.customers = {}
    def add_customer(self, name, email, phone):
        self.customers[name] = Customer(name, email, phone)
        print("Customer added")
    def add_log(self, name, message):
        if name in self.customers:
            self.customers[name].logs.append(message)
            print("Log added")
        else:
            print("Customer not found")
    def add_sale(self, name, deal, stage):
        if name in self.customers:
            self.customers[name].sales.append({"deal": deal, "stage": stage})
            print("Sale added")
        else:
            print("Customer not found")
    def update_stage(self, name, deal, stage):
        if name in self.customers:
            for s in self.customers[name].sales:
                if s["deal"] == deal:
                    s["stage"] = stage
                    print("Stage updated")
                    return
            print("Deal not found")
        else:
            print("Customer not found")
    def show_customer(self, name):
        if name in self.customers:
            c = self.customers[name]
            print("Name:", c.name)
            print("Email:", c.email)
            print("Phone:", c.phone)
            print("Logs:", c.logs)
            print("Sales:", c.sales)
        else:
            print("Customer not found")
crm = CRM()
while True:
    print("\n1 Add Customer")
    print("2 Add Communication Log")
    print("3 Add Sale")
    print("4 Update Sale Stage")
    print("5 Show Customer")
    print("6 Exit")
    ch = input("Enter choice: ")
    if ch == "1":
        n = input("Name: ")
        e = input("Email: ")
        p = input("Phone: ")
        crm.add_customer(n, e, p)
    elif ch == "2":
        n = input("Customer name: ")
        msg = input("Message: ")
        crm.add_log(n, msg)
    elif ch == "3":
        n = input("Customer name: ")
        d = input("Deal name: ")
        s = input("Stage: ")
        crm.add_sale(n, d, s)
    elif ch == "4":
        n = input("Customer name: ")
        d = input("Deal name: ")
        s = input("New stage: ")
        crm.update_stage(n, d, s)
    elif ch == "5":
        n = input("Customer name: ")
        crm.show_customer(n)
    elif ch == "6":
        break
    else:
        print("Invalid choice")
```
